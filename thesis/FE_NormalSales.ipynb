```python
Normal_sales_data=pd.read_csv(r"C:\Users\ss\OneDrive\Desktop\Msc Project\Datasets\Normal_sales_data.csv")
Normal_sales_data['DATE'] = pd.to_datetime(Normal_sales_data['DATE'])
Normal_sales_data
```

```python
Normal_sales_data['YEAR'] = Normal_sales_data['DATE'].dt.year
Normal_sales_data['MONTH'] = Normal_sales_data['DATE'].dt.month
Normal_sales_data['QUARTER'] = Normal_sales_data['DATE'].dt.quarter

Normal_sales_data['day_of_week']= Normal_sales_data['DATE'].dt.weekday
Normal_sales_data['week_of_year']= Normal_sales_data['DATE'].dt.isocalendar().week
```

```python
# Example for 1-month lag
Normal_sales_data['SALES_LAG_1'] = Normal_sales_data.groupby('ITEM DESCRIPTION')['TOTAL SALES'].shift(1)
Normal_sales_data['SALES_LAG_3'] = Normal_sales_data.groupby('ITEM DESCRIPTION')['TOTAL SALES'].shift(3)
Normal_sales_data['SALES_LAG_6'] = Normal_sales_data.groupby('ITEM DESCRIPTION')['TOTAL SALES'].shift(6)

# Example rolling average over last 3 months
Normal_sales_data['SALES_ROLLING_MEAN_3'] = Normal_sales_data.groupby('ITEM DESCRIPTION')['TOTAL SALES'].transform(lambda x: x.rolling(window=3).mean())
```

```python
# One-hot encoding for ITEM TYPE
Normal_sales_data = pd.get_dummies(Normal_sales_data, columns=['ITEM TYPE'], prefix='TYPE')

# Label encoding for SUPPLIER
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Normal_sales_data['SUPPLIER_ENCODED'] = le.fit_transform(Normal_sales_data['SUPPLIER'])
```

```python
# Median sales of each product across all periods
median_sales = Normal_sales_data.groupby('ITEM DESCRIPTION')['TOTAL SALES'].median().reset_index()
median_sales.rename(columns={'TOTAL SALES': 'MEDIAN_PRODUCT_SALES'}, inplace=True)

# Merge back
Normal_sales_data = Normal_sales_data.merge(median_sales, on='ITEM DESCRIPTION', how='left')
```

```python
# Count number of outliers per month-year
normal_data_unstack = Normal_sales_data.groupby(['YEAR', 'MONTH']).size().unstack(fill_value=0)
normal_data_unstack
```
